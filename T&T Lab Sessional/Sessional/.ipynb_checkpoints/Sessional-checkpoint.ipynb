```python
# Filter the dataset for Ford brand and train regression model

ford_data = cars_df[cars_df['brand'] == 'Ford']
ford_data

new_ford_data = ford_data.drop("brand", axis=1)
new_ford_data = new_ford_data.drop("model", axis=1)
new_ford_data = new_ford_data.drop("transmission", axis=1)
new_ford_data = new_ford_data.drop("color", axis=1)
new_ford_data = new_ford_data.drop("interior", axis=1)
new_ford_data = new_ford_data.drop("saledate", axis=1)
new_ford_data = new_ford_data.drop("body", axis=1)


new_ford_data_new = new_ford_data[new_ford_data['odometer'] < 500]
new_ford_data_new


# You can use any number of features you want.
# For testing purpose, use the data for which selling price is NULL
```

```python
# Regression model should be able to predict average selling price for a car for unknown data

from sklearn.linear_model import LogisticRegression

X = new_ford_data_new.drop("sellingprice", axis=1)
X
```

```python
y = new_ford_data_new['sellingprice']
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y)
X_test
```

```python
lr = LogisticRegression()
lr.fit(X_train, y_train)
```

```python
y_pred = lr.predict(X_test)
```

```python
y_pred
```

```python
lr.score(X_test, y_pred)
```

```python
from sklearn.metrics import confusion_matrix, classification_report
```

```python
confusion_matrix(y_test, y_pred)
```
